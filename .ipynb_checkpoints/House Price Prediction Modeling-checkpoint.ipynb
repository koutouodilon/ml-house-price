# Problem statement

For this task we're going to use data from California census to create a machine learning model to predict house prices in the State. The data includes features such as population, median income, and median house prices for each block group in California.

A block unit, also called a district, is the smallest geographic unit which typically has a population of 600 to 3,000 people. Our objective is to build a Machine Learning model from the data to predict the median house price in any neighbourhood.

It is important to have a sufficient number of instances in your dataset for each stratum, otherwise, the estimate of the importance of a stratum may be biased. This means that you should not have too many strata and that each stratum should be large enough:

```python
import numpy as np
df_housing['income_cat'] = pd.cut(df_housing['median_income'], bins=[0., 1.5, 3.0, 4.5, 6., np.inf], labels=[1, 2, 3, 4, 5])
df_housing['income_cat'].hist()
plt.show()
```

## Stratified sampling on the dataset

To increase the precision of an estimator, we need to use a sampling scheme that can reduce the heterogeneity of the population. If the data set is heterogeneous with respect to the characteristic of interest, one of these sampling procedures is stratified sampling.

In Data Science, the basic idea of stratified sampling is to:

- Divide the entire heterogeneous population into smaller groups or subpopulations such that the sampling units are homogeneous with respect to the characteristic of interest within the subpopulation.
- Treat each subpopulation as a separate population.

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df_housing, df_housing["income_cat"]):
    strat_train_set = df_housing.loc[train_index]
    strat_test_set = df_housing.loc[test_index]
print(strat_test_set['income_cat'].value_counts() / len(strat_test_set))
```

Let's remove the income_cat feature we added before.

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop('income_cat', axis=1, inplace=True)
df_housing = strat_train_set.copy()
```

# Data Preparation

Let's drop the median house value from the training set

```python
df_housing = strat_train_set.drop("median_house_value", axis=1)
```

Let's get the prediction target (median house value)

```python
df_housing_labels = strat_train_set["median_house_value"].copy()
```

We're going to replace the total bedrooms missing values with the median total bedrooms

```python
median = df_housing["total_bedrooms"].median()
df_housing["total_bedrooms"].fillna(median, inplace=True)
```

```python
housing_num = df_housing.drop("ocean_proximity", axis=1)
```

```python
df_housing.head()
```

```python
# column index
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6
```

Let's create a transformer to automatically add the rooms per household, population per household and bedrooms per room

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room

    def fit(self, X, y=None):
        return self  # nothing else to do

    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
housing_num_tr
```

```python
from sklearn.compose import ColumnTransformer
```

```python
num_attribs = list(housing_num)
```

```python
num_attribs
```

```python
cat_attribs = ["ocean_proximity"]
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])
df_housing_prepared = full_pipeline.fit_transform(df_housing)
```

# Training & Evaluation

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
lin_reg = LinearRegression()
lin_reg.fit(df_housing_prepared, df_housing_labels)
```

```python
df_housing.info()
```

```python
strat_test_set.info()
```

```python
X_test = strat_test_set.drop('median_house_value', axis=1)
y_test = strat_test_set['median_house_value'].copy()
```

```python
X_test_prepared = full_pipeline.transform(X_test)
```

```python
y_predicted = lin_reg.predict(X_test_prepared)
```

```python
mae = mean_absolute_error(y_test, y_predicted)
mse = mean_squared_error(y_test, y_predicted)
r2 = r2_score(y_test, y_predicted)

print("The model performance for testing set")
print("--------------------------------------")
print('MAE is {}'.format(mae))
print('MSE is {}'.format(mse))
print('R2 score is {}'.format(r2))
```
